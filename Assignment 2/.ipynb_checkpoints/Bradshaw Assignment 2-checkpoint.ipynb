The plot should have the x-axis labeled ($\omega$ in radians) and the average $|X|$ value for that $\omega$ on the y-axis.</li>

```python
%matplotlib inline
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import warnings
plt.style.use('seaborn-notebook') # awesome

def rms_error(x, x_hat):
    return np.sqrt(np.mean(np.squeeze(np.asarray(x-xhat))**2))

def rms_power(x_t):
    return np.sqrt(np.mean(np.square(x_t)))

def generate_signal(T, dt, rms, limit, seed=None):
    if seed:
        np.random.seed(seed) # make random numbers predicatable 
    
    nSamples = int(T/dt)
    
    # generate the positive frequency components
    X = np.zeros(nSamples//2, dtype=np.complex)
    real_samples = np.random.normal(size=nSamples//2)
    imag_samples = np.random.normal(size=nSamples//2)
    
    X.real = real_samples
    X.imag = imag_samples
    
    # eliminate frequency components greater than limit
    set_to_zero = np.fft.rfftfreq(nSamples-1, dt) > limit
    X[set_to_zero] = 0
 
    # we want X(w) = X(-w)* so use irfft to simplify the process
    x = np.real(np.fft.irfft(X, nSamples))
    rms_x = rms_power(x)
    rescaled_x = x*(rms/rms_x) # rescale the time domain signal's RMS
    
    # this is cludgy. I used rfft for convenience, but I can't figure out
    # how to plot it, so I take the conventional fourier transform.
    return rescaled_x, np.fft.fft(rescaled_x)

T = 1
dt = 0.001
N = int(T/dt)
rms = 0.5
time = np.linspace(0, T, N)

# limits 5, 10, 20
x5, _ = generate_signal(T, dt, rms, 5, seed=15)
x10, _ = generate_signal(T, dt, rms, 10, seed=16)
x20, _ = generate_signal(T, dt, rms, 20, seed=17)

dw = 2*np.pi/T
limit = 10
time = np.linspace(0, T, N)
freq = np.array([dw*n if n<N/2 else dw*(n-N) for n in range(N)])

fourier_coefficients = np.zeros((N, 100))
for ii in xrange(0, 100):
    _, X = generate_signal(T, dt, rms, limit, seed=ii)
    fourier_coefficients[:, ii]  = np.abs(X)

mean_fourier_coeffients = np.mean(fourier_coefficients, axis=1)

# 1.1a figure
plt.figure()
plt.plot(time, x5, label='Limit=5')
plt.plot(time, x10, label='Limit=10')
plt.plot(time, x20, label='Limit=20')
plt.xlabel('time (seconds)')
plt.ylabel('Arbitray Units')
plt.title('1.1A White Noise Signals with Various Frequency Cutoffs')
plt.legend()

# 1.1b figure
plt.figure()
plt.plot(np.fft.fftshift(freq), np.fft.fftshift(mean_fourier_coeffients))
plt.xlabel('Freqency in rads/s')
plt.ylabel('Average of Fourier Coeffients (|X(w)|)')
plt.title('1.1B: Fourier Spectrum of the White Noise Signal')
plt.show()

# 1.1b figure zoomed in
plt.figure()
plt.plot(np.fft.fftshift(freq), np.fft.fftshift(mean_fourier_coeffients))
plt.xlabel('Freqency in rads/s')
plt.ylabel('Average of Fourier Coeffients (|X(w)|)')
plt.title('1.1B: Fourier Spectrum of the White Noise Signal (zoomed-in)')
plt.xlim([-100, 100])
plt.show()
```

### 1.2) Gaussian power spectrum noise

<ol style="list-style:lower-alpha outside none;" start="1">
<li>[1 mark] Plot $x(t)$ for three randomly generated signals with ``bandwidth`` at 5, 10, and 20Hz.  For each of these, ``T``=1, ``dt``=0.001, and ``rms``=0.5.</li>

<li>[1 mark] Plot the average $|X(\omega)|$ (the norm of the Fourier coefficients) over 100 signals generated with  ``T``=1, ``dt``=0.001, ``rms``=0.5, and ``bandwidth``=10 (each of these 100 signals should have a different ``seed``).</li>
</ol>

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt

def gaussian_with_bandwidth(w, b):
    sigma = np.exp(-w**2/(2*b**2))
    # keep output about python double's minimum float
    if sigma > 0:
        return np.random.normal(loc=0, scale=sigma)
    else:
        return 0

def generate_gaussian_ps_noise(b, T, N):
    ps_noise = np.zeros(N)
    dw = 2*np.pi / T
    
    for ii in xrange(N):
        w = ii * dw
        ps_noise[ii] = gaussian_with_bandwidth(w, b)
    return ps_noise

def generate_signal_gaussian_ps(T, dt, rms, limit, seed=None):
    if seed:
        np.random.seed(seed) # make random numbers predicatable 
        
    # I am using python's rfft method
    # rfft will take care of X(w) = X(-w)* for me
     
    # cycles to radians The units of b were not explicitely stated
    # but to me it makes sense that we should try to generate the same
    # frequencies as we did in part 1.1a so that the plots are comparable
    bandwidth_limit = limit*np.pi*2
    N = int(T/dt)
    # generate the positive frequency components
    
    generated_coeffs = np.zeros(N//2, dtype=np.complex)
    # remember that delta w = 2pi / T
    real_samples = generate_gaussian_ps_noise(bandwidth_limit, T, N//2)
    imag_samples = generate_gaussian_ps_noise(bandwidth_limit, T, N//2)
    
    generated_coeffs.real = real_samples
    generated_coeffs.imag = imag_samples
 
    # X(w) = X(w)* so use irfft
    x = np.real(np.fft.irfft(generated_coeffs, N))
    rms_x = rms_power(x)
    rescaled_x = x*(rms/rms_x)
    return rescaled_x, np.fft.fft(rescaled_x)

N = int(T/dt)
T = 1
dt = 0.001
rms = 0.5
time = np.linspace(0, T, N)
dw = 2*np.pi/T
freq = np.array([dw*n if n<N/2 else dw*(n-N) for n in range(N)])

# 1.2 A
# limits 5, 10, 20
x5, _ = generate_signal_gaussian_ps(T, dt, rms, 5, seed=1)
x10, _ = generate_signal_gaussian_ps(T, dt, rms, 10, seed=2)
x20, X = generate_signal_gaussian_ps(T, dt, rms, 20, seed=3)

# 1.2B
limit = 10
time = np.linspace(0, T, N)

fourier_coefficients = np.zeros((N, 100))
for ii in xrange(0, 100):
    _, X = generate_signal_gaussian_ps(T, dt, rms, limit, seed=ii)
    fourier_coefficients[:, ii]  = np.abs(X)

mean_fourier_coeffients = np.mean(fourier_coefficients, axis=1)

# 1.2a figure
plt.figure()
plt.plot(time, x5, label='Limit=5')
plt.plot(time, x10, label='Limit=10')
plt.plot(time, x20, label='Limit=20')
plt.title('White Noise Gaussian Bandwidth Limited Signals')
plt.xlabel('time (seconds)')
plt.ylabel('Arbitray Units')
plt.legend()

# 1.2b figure
plt.figure()
plt.plot(np.fft.fftshift(freq), np.fft.fftshift(mean_fourier_coeffients))
plt.xlabel('Freqency in rads/s')
plt.ylabel('Average of Fourier Coeffients (|X(w)|)')
plt.title('1.2B: Fourier Spectrum of the White Noise Signal')
plt.show()

# 1.2b figure zoomed in
plt.figure()
plt.plot(np.fft.fftshift(freq), np.fft.fftshift(mean_fourier_coeffients))
plt.xlabel('Freqency in rads/s')
plt.ylabel('Average of Fourier Coeffients (|X(w)|)')
plt.title('1.2B: Average Fourier Spectrum of the White Noise Signal (zoomed-in)')
plt.xlim([-200, 200])
plt.show()
```

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt

def lif_step(tau_rc, J, V):
    return (J-V)/tau_rc

def lif_neuron(x_values, tau_ref, tau_rc, e, gain, bias):
    J = e * x_values * gain + bias
    neuron_val = []
    for j in J:
        if j > 1:
            neuron_val.append(1 / (tau_ref - tau_rc*np.log(1-1/j)))
        else:
            neuron_val.append(0)
    return np.array(neuron_val)

def generate_J(x_values, gain, encoder, bias):
    return gain * encoder * x_values + bias

def spiking_neuron(tau_rc, tau_ref, J_values, dt):
    """this method solves the ODE and all the special cases related to spiking"""
    
    # constants
    SPIKE = 2 # spikes denoted by V=2
    REFRACTORY = 0 # refractory period V=0
    ref_period = int(tau_ref / dt)
    
    ref_count = 0
    nSpikes = 0
    refractory = False
    
    V_values = np.zeros(np.size(J_values))
    V = 0
    # euler's method
    for ii, jv in enumerate(J_values):
        # reset voltage
        # incrememnt the refractory period if necessary
        if refractory and ref_count < ref_period:
            V_values[ii] = REFRACTORY
            ref_count += 1
        elif refractory:
            refractory = False
            ref_count = 0
            V += dt*lif_step(tau_rc, jv, V)
        else:
            V += dt*lif_step(tau_rc, jv, V)
        
        # reset voltage if it drops below zero
        if V < 0:
            V = 0
            
        # spike if necessary
        if V>=1:
            nSpikes += 1
            V_values[ii] = SPIKE
            V = 0
            refractory = True
        else:
            V_values[ii] = V 
    return V_values, nSpikes

def binary_spiking_neuron(tau_rc, tau_ref, J_values, dt):
    """this method solves the ODE and all the special cases related to spiking
    
    unlike spiking_neuron, this method returns only the spike locations. Spiking
    neuron returns the analog voltage, which I found very useful for debugging.
    """
    V_values, nSpikes = spiking_neuron(tau_rc, tau_ref, J_values, dt)
    spike_locations = np.zeros(np.size(V_values))
    spike_locations[V_values==2] = 1
    return spike_locations, nSpikes
    
tau_rc = 0.02
tau_ref = 0.002
T = 1
dt = 0.001
N = int(T/dt)
time_values = np.linspace(0, T, N)
# calculated using the LIF equation
GAIN = 3.345
BIAS = 1.46
encoder = 1
# 2a just checking my math by generating the rate curve
lif_vals = lif_neuron(time_values, tau_ref, tau_rc, 1, GAIN, BIAS)

# 2a
X0 = np.zeros(N)
JX0 = generate_J(X0, GAIN, encoder, BIAS)
X0spiking_neuron_vals, nSpikes = spiking_neuron(tau_rc, tau_ref, JX0, dt)
print "Number of Spikes with constant x(t)=0: {}".format(nSpikes)
# 2b
X1 = np.ones(N)
JX1 = generate_J(X1, GAIN, encoder, BIAS)
X1spiking_neuron_vals, nSpikes = spiking_neuron(tau_rc, tau_ref, JX1, dt)
print "Number of Spikes with constant x(t)=1: {}".format(nSpikes)
# 2c
x_t, _ = generate_signal(1, 0.001, 0.5, 30, seed=ii)
J_x_t = generate_J(x_t, GAIN, encoder, BIAS)
Xtspiking_neuron_vals, nSpikes = spiking_neuron(tau_rc, tau_ref, J_x_t, dt)
print "Number of Spikes with randomly generated signal: {}".format(nSpikes)

#2a LIF RATE plot (not specifically requested, but seems worth including)
plt.figure()
plt.plot(time_values, lif_vals)
plt.title('LIF Firing Rates')
plt.xlabel('x')
plt.ylabel('2A: Firing Rate (Hz)')
plt.show()

# 2a figure
plt.figure(figsize=(14,3))
plt.plot(time_values, X0)
plt.vlines(time_values[np.where(X0spiking_neuron_vals==2)],
           -1, 1, 'k', label='Spikes', alpha=0.5)
plt.title('2A: LIF Neuron with x(t)=0')
plt.xlabel('')
plt.ylim([-1.1,1.1])
plt.show()

# 2b figure
plt.figure(figsize=(14, 3))
plt.plot(time_values, X1)
plt.vlines(time_values[np.where(X1spiking_neuron_vals==2)],
           -1, 1, 'k', label='Spikes', alpha=0.5)
plt.title('2B: LIF Neuron with x(t)=1')
plt.xlabel('')
plt.ylim([-1.1,1.1])
plt.show()

# 2c figure
plt.figure(figsize=(14, 3))
plt.plot(time_values, x_t, 'r-', label='x(t)')
plt.vlines(time_values[np.where(Xtspiking_neuron_vals==2)],
           -1, 1, 'k', label='Spikes', alpha=0.5)
plt.title('2C: Spike Pattern with x(t)')
plt.xlabel('')
plt.legend()
plt.show()

# 2d figure (just a scaled version of the previous)
plt.figure(figsize=(14, 3))
plt.plot(time_values, x_t, 'r-', label='x(t)')
plt.plot(time_values, Xtspiking_neuron_vals, 'b', label='Spiking Neuron')
plt.vlines(time_values[np.where(Xtspiking_neuron_vals==2)],
           -1, 2, 'k', label='Spikes', alpha=0.5)
plt.title('2D: Spike Pattern with x(t)')
plt.xlabel('')
plt.xlim([0, 0.2])
plt.legend()
plt.show()
```

```python
import numpy

# TODO: change this to use the generate_signal from 1.1

T = 4.0         # length of signal in seconds
dt = 0.001      # time step size

# Generate bandlimited white noise (use your own function from part 1.1)
x, X = generate_signal(T, dt, rms=0.5, limit=5, seed=3)

X = numpy.fft.fftshift(X)

Nt = len(x)        # number of time points to simulate the two neurons over
t = numpy.arange(Nt) * dt  # list of simulation time points

# Neuron parameters
tau_ref = 0.002   # neuron refractory period. in this case (2ms or 2*dt)
tau_rc = 0.02     # LIF neuron RC time constant
x0 = 0.0          # firing rate at x=x0 is a0
a0 = 40.0
x1 = 1.0          # firing rate at x=x1 is a1
a1 = 150.0

# calculate neuron gain and bias
eps = tau_rc/tau_ref
r1 = 1.0 / (tau_ref * a0)
r2 = 1.0 / (tau_ref * a1)
f1 = (r1 - 1) / eps
f2 = (r2 - 1) / eps
alpha = (1.0/(numpy.exp(f2)-1) - 1.0/(numpy.exp(f1)-1))/(x1-x0) 
x_threshold = x0-1/(alpha*(numpy.exp(f1)-1))              
Jbias = 1-alpha*x_threshold;   

# Simulate the two neurons (use your own function from part 3)
spikes, _, _ = two_neurons(x, dt, alpha, Jbias, tau_rc, tau_ref)

freq = numpy.arange(Nt)/T - Nt/(2.0*T)   # an array of frequencies (dw)
        # that correspond to fft signals (after fftshift)
    
omega = freq*2*numpy.pi     # convert frequency array from Hz to radians

r = spikes[0] - spikes[1]   # compute the difference of spike trains 
R = numpy.fft.fftshift(numpy.fft.fft(r)) # fft of spike train signal

sigma_t = 0.025                          # standard deviation of the gaussian window
W2 = numpy.exp(-omega**2*sigma_t**2)     # a gaussian window in the frequency domain
W2 = W2 / sum(W2)                        # normalize the window

CP = X*R.conjugate() # numerator of optimal filter equation in the frequency domain
WCP = numpy.convolve(CP, W2, 'same')  # apply the gaussian window to CP

RP = R*R.conjugate()       # magnitude of R (R is complex so |R|=R*R*) 
                           #  denomonator optimal filter in freq domain
    
WRP = numpy.convolve(RP, W2, 'same')  # apply the gaussian window to RP
XP = X*X.conjugate()                  # magnitude of X (R is complex so |X|=X*X*)
WXP = numpy.convolve(XP, W2, 'same')  # apply the gaussian window to XP

# compute the optimal filter equation derived in the temporal rep lecture
H = WCP / WRP

# convert the optimal filter to time domain
h = numpy.fft.fftshift(numpy.fft.ifft(numpy.fft.ifftshift(H))).real  

XHAT = H*R  # decode the spike train using the optimal filter

# transform the filter output back into the time domain
xhat = numpy.fft.ifft(numpy.fft.ifftshift(XHAT)).real 

print "RMSE: {}".format(rms_error(x, xhat))

plt.figure(figsize=(14, 4))
plt.subplot(1,2,1)
# plot vector norm of input signal in frequency domain
plt.plot(freq, numpy.sqrt(XP), label='$|X(\omega)|$')  
plt.legend()
plt.xlabel('w')
plt.ylabel('$|X(\omega)|$')
plt.title('4A: $|X(\omega)|$ Power Spectrum')

plt.subplot(1,2,2)
# plot vector norm of difference of spike trains in the frequency domain
plt.plot(freq, numpy.sqrt(RP), label='$|R(\omega)|$')  
plt.legend()
plt.xlabel('w')
plt.ylabel('$|R(\omega)|$')
plt.title('4A: $|R(\omega)|$ Power Spectrum')

plt.figure(figsize=(14, 4))
plt.subplot(1,2,1)
plt.plot(freq, H.real) # Real part of the ideal filter in the frequency domain
plt.xlabel('w')
plt.title('')
plt.title('4A: Optimal Filter in Frequency Domain')
plt.xlim(-50, 50)

plt.subplot(1,2,2)
plt.plot(t-T/2, h)       # plot optimal filter in the time domain
plt.title('4A: Optimal Filter in Time Domain')
plt.xlabel('t')
plt.ylabel('h(t)')
plt.xlim(-0.5, 0.5)

plt.figure(figsize=(14, 4))
plt.plot(t, r, color='k', label='Difference of Spikes', alpha=0.2) # difference of spikes
plt.plot(t, x, linewidth=2, label='White Signal')   # white signal
plt.plot(t, xhat, label='Filter Output')            # filter output
plt.title('4A: Filter Results')
plt.legend(loc='best')
plt.xlabel('t')
plt.show()
```

<p>
<b>Answer:</b> The filter becomes narrower in the time domain, and its peak magnitude increases. This means that the filter performs less temporal smoothing on the spike train. This happens because the filter cannot smooth the high frequency signal's spike trains very much without eliminating the high frequency signal components. The results obtained become less accurate at higher frequencies, because the representational ability of only two neurons is limited. The RMSE values listed below refect this.
</p>
</ol>


```python
def compute_optimal_filter(T, dt, x, X, spikes, freq_limit):
    X = numpy.fft.fftshift(X)

    Nt = len(x)                
    t = numpy.arange(Nt) * dt  

    freq = numpy.arange(Nt)/T - Nt/(2.0*T)   
    omega = freq*2*numpy.pi                  

    r = spikes[0] - spikes[1]                
    R = numpy.fft.fftshift(numpy.fft.fft(r)) 

    sigma_t = 0.025                          
    W2 = numpy.exp(-omega**2*sigma_t**2)     
    W2 = W2 / sum(W2)                        

    CP = X*R.conjugate()                  
    WCP = numpy.convolve(CP, W2, 'same')  
    RP = R*R.conjugate()                  
    WRP = numpy.convolve(RP, W2, 'same')  
    XP = X*X.conjugate()                  
    WXP = numpy.convolve(XP, W2, 'same')  

    H = WCP / WRP                         

    h = numpy.fft.fftshift(numpy.fft.ifft(numpy.fft.ifftshift(H))).real  

    XHAT = H*R                            

    xhat = numpy.fft.ifft(numpy.fft.ifftshift(XHAT)).real  

    print "RMSE with limit={}: {}".format(freq_limit, rms_error(x, xhat))
    
    plt.figure(figsize=(14, 4))
    plt.plot(t-T/2, h)       
    plt.title('Optimal Filter in Time Domain for Frequency Limit={}'.format(freq_limit))
    plt.xlabel('t')
    plt.ylabel('h(t)')
    plt.xlim(-0.5, 0.5)

    plt.figure(figsize=(14, 4))
    plt.plot(t, r, color='k', label='Difference of Spikes', alpha=0.2)  
    plt.plot(t, np.zeros(np.size(t)), 'w', linewidth=2)
    plt.plot(t, x, linewidth=2, label='White Signal')           
    plt.plot(t, xhat, label='Filter Output')                     
    plt.title('Filter Results in the time domain for Frequency Limit={}'.format(freq_limit))
    plt.legend(loc='best')
    plt.xlabel('t')
    plt.show()

# 4e
T = 2
dt = 0.001
rms = 0.5
GAIN = 3.345
BIAS = 1.46
tau_rc = 0.02
tau_ref = 0.002
# limit 2
x2, X2 = generate_signal(T, dt, rms, 2, seed=12)
spikes5, _, _ = two_neurons(x2, dt, GAIN, BIAS, tau_rc, tau_ref)
compute_optimal_filter(T, dt, x2, X2, spikes5, 5)
# limit 10
x10, X10 = generate_signal(T, dt, rms, 10, seed=21)
spikes10, _, _ = two_neurons(x10, dt, GAIN, BIAS, tau_rc, tau_ref)
compute_optimal_filter(T, dt, x10, X10, spikes10, 10)
# limit 30
x30, X30 = generate_signal(T, dt, rms, 20, seed=121)
spikes30, _, _ = two_neurons(x30, dt, GAIN, BIAS, tau_rc, tau_ref)
compute_optimal_filter(T, dt, x30, X30, spikes30, 30)
```

## 5) Using Post-Synaptic Currents as a Filter

Instead of using the optimal filter from the previous question, now we will use the post-synaptic current instead.  This is of the form $h(t)=t^n e^{-t/\tau}$ normalized to area 1.

<ol style="list-style:lower-alpha outside none;" start="1">
<li>[1 mark] Plot the normalized $h(t)$ for $n$=0, 1, and 2 with $\tau$=0.007 seconds.  What two things do you expect increasing $n$ will do to $\hat{x}(t)$?</li>

<p>
**Answer:** I expect that increasing $n$ will cause $\hat{x}(t)$ to be more temporally smoothed, and that increasing $n$ will increase the phase delay of the filter. The temporal smoothing will increase, because increasing $n$ widens $h(t)$. The phase delay will increase because $h(t)$ is a causal filter, and its output will depend more on past input when $h(t)$ is widened.

**Answer:** The signal that the decoders were computed on was decoded more accurately than the second 2Hz signal (RMSE 0.12 vs RMSE 0.86). It makes sense that the signal that the decoders were calculated on would be decoded most accurately, especially considering the limited representational power of a population with only two neurons.
</p>

```python
T = 2
dt = 0.001
N = int(T/dt)
noise_std = 0.2
t_A = np.linspace(0, T, N)

def generate_h_t(t_h, n, tau):
    h = t_h**n * np.exp(-t_h/tau)
    h[np.where(t_h<0)]=0
    h = h/np.linalg.norm(h,1) # normalize to area 1
    return h

def generate_decoders(A, X, dt, noise_std):
    Ypsilon = dt * np.matmul(A.T,X)
    Gamma = dt * np.matmul(A.T, A)
    Gamma_inv = np.linalg.pinv(Gamma)
    d = np.dot(Gamma_inv, Ypsilon)
    return d

#5a 
t_h = np.linspace(-T/2, T/2, N)
tau = 0.007
h_t_n0 = generate_h_t(t_h, 0, tau)
h_t_n1 = generate_h_t(t_h, 1, tau)
h_t_n2 = generate_h_t(t_h, 2, tau)

#5b
n = 0
h_t_t0 = generate_h_t(t_h, 0, 0.002)
h_t_t1 = generate_h_t(t_h, 1, 0.005)
h_t_t2 = generate_h_t(t_h, 2, 0.01)
h_t_t3 = generate_h_t(t_h, 2, 0.02)

# 5c
# spikes from 3D
# convolve spikes with filter and populate A matrix with the result
h_t = generate_h_t(t_h, 0, 0.007)

fspikes15c = np.convolve(spikes3d[0], h_t, mode='same')
fspikes25c = np.convolve(spikes3d[1], h_t, mode='same')

A5c = np.array([fspikes15c, fspikes25c]).T
d5c = generate_decoders(A5c, x_t2, dt, noise_std)
xhat5c = np.dot(A5c, d5c)

# difference of spikes for the plot
diff_spikes3d = spikes3d[0] - spikes3d[1]

# generate frequency plot H(w)
dw = 2*np.pi/T
freq = np.array([dw*n if n<N/2 else dw*(n-N) for n in range(N)])
H_w_n0 = np.fft.fft(h_t_n0)

# 5d
T = 2
dt = 0.001
rms = 0.5
limit = 2
N = int(T/dt)

x_t25d, _ = generate_signal(T, dt, rms, limit, seed=21)
spikes5d, nspikes1, nspikes2 = two_neurons(x_t25d, dt, GAIN, BIAS, tau_rc, tau_ref)
print "Number of Spikes with white signal T=2, dt=0.001, rms=0.5, and limit=5: {}".format(nspikes1)
print "Number of Spikes with white signal T=2, dt=0.001, rms=0.5, and limit=5: {}".format(nspikes2)

fspikes1_5d = np.convolve(spikes5d[0], h_t, mode='same')
fspikes2_5d = np.convolve(spikes5d[1], h_t, mode='same')

A5d = np.array([fspikes1_5d, fspikes2_5d]).T
xhat5d = np.dot(A5d, d5c)

diff_spikes5d = spikes5d[0] - spikes5d[1]

# 5a figure
plt.figure(figsize=(14,3))
plt.plot(t_h, h_t_n0, label='n=0')
plt.plot(t_h, h_t_n1, label='n=1')
plt.plot(t_h, h_t_n2, label='n=2')
plt.title('5A: h(t) for n=0, n=1, n=2')
plt.xlabel('t')
plt.ylabel('h(t)')
plt.xlim([-.2, 0.2])
plt.legend()
plt.show()

# 5b figure
plt.figure(figsize=(14,3))
plt.plot(t_h, h_t_t0, label='tau=0.002')
plt.plot(t_h, h_t_t1, label='tau=0.005')
plt.plot(t_h, h_t_t2, label='tau=0.01')
plt.plot(t_h, h_t_t3, label='tau=0.02')
plt.title('5B: h(t) for tau=0.002, tau=0.005, tau=0.01, tau=0.02')
plt.xlabel('t')
plt.ylabel('h(t)')
plt.xlim([-.2, 0.2])
plt.legend()
plt.show()

#5c figures
plt.figure(figsize=(14,3))
plt.plot(t_h, h_t_n0)
plt.ylabel('h(t)')
plt.xlabel('t')
plt.title('5C: Time domain h(t) plot')
plt.xlim([-.2, 0.2])
plt.figure(figsize=(14,3))
plt.plot(np.fft.fftshift(freq), np.abs(np.fft.fftshift(H_w_n0)))
plt.ylabel('H(w)')
plt.xlabel('w')
plt.title('Frequency Domain H(w) plot')
plt.show()

print "5C RMSE Decoding signals: {}".format(rms_error(x_t2, xhat5c))

plt.figure(figsize=(14,3))
plt.plot(t_A, diff_spikes3d, 'k', label='Spikes', alpha=0.2)
plt.plot(t_A, xhat5c, label='xhat')
plt.plot(t_A, x_t2, 'r', label='x(t)')
plt.ylabel('H(w)')
plt.xlabel('w')
plt.legend()
plt.title('5C x(t), spikes and decoded signal')
plt.show()

#5d figure
print "5D RMSE Decoding signals: {}".format(rms_error(x_t25d, xhat5d))

plt.figure(figsize=(14,3))
plt.plot(t_A, diff_spikes5d, 'k', label='Spikes', alpha=0.2)
plt.plot(t_A, xhat5d, label='xhat')
plt.plot(t_A, x_t25d, 'r', label='x(t)')
plt.ylabel('')
plt.xlabel('')
plt.legend()
plt.title('5D x(t) and filtered spikes')
plt.show()
```
